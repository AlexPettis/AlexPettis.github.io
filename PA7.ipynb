```python
import pandas as pd

df = pd.read_csv('Boone.csv')

df.info()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])

df
```

```python
df = df[df['placename'] == 'Boone'].drop('placename', axis = 1)
df = df[df['usaf'] == 722198].drop('usaf', axis = 1)
df = df[df['wban'] == 63819].drop('wban', axis = 1)
df = df.drop('pressure', axis = 1)

df
```

```python
df = df.rename(columns = {'air_temp':'AirTemp', 'dew_point':'DewPoint', 'wind_dir':'WindDiretion', 'wind_spd':'WindSpeed','sky_cond':'SkyCond', 'datetime': 'DateTime', 'precip_1hr':'Precip_1HR'})

df
```

```python
missing_data_percent = (df.isna().sum() / len(df)) * 100
print(missing_data_percent)
```

```python
round(df.corr(),2)
```

```python
df.to_csv('boone_clean.csv', index = False)
```
